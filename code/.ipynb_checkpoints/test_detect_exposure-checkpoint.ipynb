## 分類通過與沒通過

```python
import os
import cv2 as cv
import pandas as pd
import numpy as np
from detect_exposure import DetectExposure
```

```python
directory_name = '../image'
df = pd.DataFrame(columns = ['image_path', 'image', 'type'])
for filename in os.listdir(r"./" + directory_name):
    next_directory_name = '../image/' + str(filename)
    for next_filename in os.listdir(r"./" +  next_directory_name):
        df.loc[len(df)] = ['../image/' + str(filename) + '/' + str(next_filename), next_filename.replace('.jpg',''), str(filename)]
df['is_blurry'] = df['image'].apply(lambda x: False if x[-1:]!='b' else True)
#排除模糊圖片
df = df[~df['is_blurry']]
df = df.reset_index(drop = True)
df = df.drop('is_blurry',axis = 1)
```

```python
for i in range(len(df)):
    image_path = df.loc[i]['image_path']
    img = cv.imread(image_path, cv.IMREAD_GRAYSCALE)
    de = DetectExposure(img)
    if de.fit():
        cv.imwrite('./pass/' + df.loc[i]['image']+'.jpg',img)
    else:
        cv.imwrite('./nopass/' + df.loc[i]['image']+'.jpg',img)
```

```python
import os
import cv2 as cv
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from detect_exposure import DetectExposure
```

```python
directory_name = '../image'
df = pd.DataFrame(columns = ['image_path', 'image', 'type'])
for filename in os.listdir(r"./" + directory_name):
    next_directory_name = '../image/' + str(filename)
    for next_filename in os.listdir(r"./" +  next_directory_name):
        df.loc[len(df)] = ['../image/' + str(filename) + '/' + str(next_filename), next_filename.replace('.jpg',''), str(filename)]
df['is_blurry'] = df['image'].apply(lambda x: False if x[-1:]!='b' else True)
#排除模糊圖片
df = df[~df['is_blurry']]
df = df.reset_index(drop = True)
df = df.drop('is_blurry',axis = 1)
#設定通過與沒通過
df['pass'] = list(pd.read_csv('img_t.csv',header=None)[0])
```

```python
amount_of_different_light_list = list()
ratio_of_background_to_image_list = list()
ratio_of_gray_to_image_list = list()
cal_pass_list = list()
for i in range(len(df)):
    image_path = df.loc[i]['image_path']
    img = cv.imread(image_path, cv.IMREAD_GRAYSCALE)
    de = DetectExposure(img)
    cal_pass_list.append(de.fit())
    amount_of_different_light_list.append(de.amount_of_different_light)
    ratio_of_background_to_image_list.append(de.ratio_of_background_to_image)
    ratio_of_gray_to_image_list.append(de.ratio_of_gray_to_image)
df['amount_of_different_light'] = amount_of_different_light_list
df['ratio_of_background_to_image'] = ratio_of_background_to_image_list
df['ratio_of_gray_to_image'] = ratio_of_gray_to_image_list
df['cal_pass'] = cal_pass_list
```

```python
#去除因光影差異而沒通過的圖片
LD_THRESHOLD = 90
df['amount_of_different_light_pass'] = df['amount_of_different_light'].apply(lambda x: False if x >= LD_THRESHOLD else True)
df_PLD = df[df['amount_of_different_light_pass']].reset_index(drop = True)
```

```python
#整體
df_pass = df_PLD[df_PLD['pass']==1]
df_nopass = df_PLD[df_PLD['pass']==0]
plt.figure(figsize=(9,9))
plt.scatter(df_pass['ratio_of_background_to_image'], df_pass['ratio_of_gray_to_image'], label='pass')
plt.scatter(df_nopass['ratio_of_background_to_image'], df_nopass['ratio_of_gray_to_image'], label='no_pass')

my_x_ticks = np.arange(0, 0.7, 0.1)
my_y_ticks = np.arange(0, 0.7, 0.1)
plt.xticks(my_x_ticks)
plt.yticks(my_y_ticks)


WL_ALL_BACKGROUND_THRESHOLD = 0.16
WL_ALL_GRAY_THRESHOLD = 0.136
plt.plot([WL_ALL_BACKGROUND_THRESHOLD, 0.7], [WL_ALL_GRAY_THRESHOLD, WL_ALL_GRAY_THRESHOLD],color = 'red')
plt.plot([WL_ALL_BACKGROUND_THRESHOLD, WL_ALL_BACKGROUND_THRESHOLD], [WL_ALL_GRAY_THRESHOLD, 0.7],color = 'red')

plt.legend()
plt.show()
```

# 正確率

```python
tp = len(df[(df['pass']==1) & (df['cal_pass'])])
```

```python
fn = len(df[(df['pass']==0) & (df['cal_pass'])])
```

```python
fp = len(df[(df['pass']==1) & ~(df['cal_pass'])])
```

```python
tn = len(df[(df['pass']==0) & ~(df['cal_pass'])])
```

```python
print('precision：',end= ' ')
precision = tp / (tp+fp)
print(precision)
```

```python
print('recall：',end= ' ')
recall = tp / (tp+fn)
print(recall)
```

```python
print('accuracy：',end= ' ')
accuracy = (tp + tn)/ (tp + tn + fp + fn)
print(accuracy)
```

```python
print('f1-score：',end= ' ')
print(2 * (precision * recall) / (precision + recall))
```

```python
print(tn/(tn + fn))
```

```python
import cv2
import numpy as np
import pandas as pd
from exposure_image import ExposureImage

class DetectExposure:
    def __init__(self, input_img):
        self.__img = input_img
        
    def fit(self):
        #取得光影差異量
        __light_different = self.__detect_light_different()
        #取得文字最低亮度
        __word_light = self.__detect_word_light()
        return __light_different, __word_light
        
    #偵測光影差異量
    def __detect_light_different(self):
        #設定切割大小
        cut_size = 8
        img = self.__img.copy()
        #更新圖片型態
        img = self.__clear_img_type(img)
        #腐蝕文字
        img = cv2.dilate(img, np.ones((41,41), np.uint8))  
        #模糊圖片
        kernel_size = (41, 41)
        sigma = 9
        img = cv2.GaussianBlur(img, kernel_size, sigma)
        #切割圖片
        img_list = self.__cut_image(img, cut_size)
        #儲存運算值
        value_list = list()
        for single_img in value_list:
            value_list.append(single_img.mean())
        return max(value_list) - min(value_list)
    
    #偵測文字亮度
    def __detect_word_light(self):
        ex = ExposureImage(self.img)
        ex.fit()
        ex_img = ex.img
        dif = list()
        __range = 3
        __img_list = self.cut_image(ex_img, __range)
        for __img in __img_list:
            __min = __img.min()
            __max = __img.max()
            dif.append(__max - __min)
        return min(dif)
    
    #切割圖片
    def __cut_image(self, input_img, input_range):
        img_list = []
        for i in range(input_range):
            start_x = int(0 + i *(self.img.shape[1] / input_range))
            end_x = int((i+1) * (self.img.shape[1] / input_range))
            for j in range(input_range):
                start_y = int(0 + j * (self.img.shape[0] / input_range))
                end_y = int((j+1) * (self.img.shape[0] / input_range))
                img_list.append(input_img[start_y:end_y,start_x:end_x])
        return img_list                 
    
    #更新圖片型態
    def __clear_img_type(self, input_img):
        input_img = np.round(input_img)
        input_img = input_img.astype(np.uint8)
        return input_img
```

```python
def LD(img):
    d_img = img.copy()
    #校正圖片型態
    d_img = np.round(d_img)
    d_img = d_img.astype(np.uint8)
    #腐蝕文字
    bg_d_img = ~cv.erode(~d_img, np.ones((41,41), np.uint8))
    #模糊
    kernel_size = (41, 41);
    sigma = 9
    bg_d_img = cv.GaussianBlur(bg_d_img, kernel_size, sigma)
    #切割圖片
    d_img_rgb = cv.cvtColor(bg_d_img, cv.COLOR_GRAY2BGR)
    #文字大小
    fontScale = 4
    #文字字型
    font = cv.FONT_HERSHEY_SIMPLEX
    #文字顏色
    color = (255, 0, 0) 
    #文字寬度
    thickness = 4
    d_img_df = pd.DataFrame(columns = ['value','start_x', 'end_x', 'start_y', 'end_y'])
    input_range = 8
    for i in range(input_range):
        start_x = int(0 + i *(bg_d_img.shape[1] / input_range))
        end_x = int((i+1) * (bg_d_img.shape[1] / input_range))
        for j in range(input_range):
            start_y = int(0 + j * (bg_d_img.shape[0] / input_range))
            end_y = int((j+1) * (bg_d_img.shape[0] / input_range))  
            s_d_img = bg_d_img[start_y:end_y,start_x:end_x]
            d_img_df.loc[len(d_img_df)] = [int(s_d_img.mean()), start_x, end_x, start_y, end_y]
    return d_img_df
```

```python
def WL(img):
    d_img = img.copy()
    #校正圖片型態
    d_img = np.round(d_img)
    d_img = d_img.astype(np.uint8)
    ex = ExposureImage(d_img)
    d_img = ex.fit()
    #切割圖片
    d_img_rgb = cv.cvtColor(d_img, cv.COLOR_GRAY2BGR)
    #文字大小
    fontScale = 2
    #文字字型
    font = cv.FONT_HERSHEY_SIMPLEX
    #文字顏色
    color = (255, 0, 0) 
    #文字寬度
    thickness = 4
    d_img_df = pd.DataFrame(columns = ['value','start_x', 'end_x', 'start_y', 'end_y'])
    input_range = 20
    
    kk_list = list()
    for i in range(input_range):
        start_x = int(0 + i *(d_img.shape[1] / input_range))
        end_x = int((i+1) * (d_img.shape[1] / input_range))
        for j in range(input_range):
            start_y = int(0 + j * (d_img.shape[0] / input_range))
            end_y = int((j+1) * (d_img.shape[0] / input_range))  
            s_d_img = d_img[start_y:end_y,start_x:end_x]
            __min = s_d_img.min()
            __max = s_d_img.max()
            d_img_df.loc[len(d_img_df)] = [int(__max - __min), start_x, end_x, start_y, end_y]
        
            k_img = ~cv.erode(~s_d_img, np.ones((21, 21), np.uint8)) 
            kk = cv.absdiff(s_d_img, k_img)
            kk_list.append(kk.max())
        
    ret1, th1 = cv.threshold(np.uint8([kk_list]), 0, 255, cv.THRESH_OTSU)
    n_d_img_df = pd.DataFrame(columns = ['value','start_x', 'end_x', 'start_y', 'end_y'])
    for i in range(len(d_img_df)):
        if (d_img_df.loc[i]['value']) >=ret1:
            
            n_d_img_df.loc[len(n_d_img_df)] = d_img_df.loc[i]
    n_d_img_df = n_d_img_df.reset_index(drop = True)
    return n_d_img_df
```
